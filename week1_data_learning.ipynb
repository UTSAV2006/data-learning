## Days 1-3: SQL Fundamentals
SQL (Structured Query Language) is used to query and manipulate data stored in relational databases.  
These were practiced on SQLZoo — the world table contains columns: name, continent, area, population, gdp.

### The core SQL skeleton — memorize this
```sql
SELECT [what columns you want]
FROM [which table]
WHERE [your filter condition]
```
Every SQL query you write will follow this structure.

### Day 1 — SELECT, WHERE, LIKE

**Basic SELECT — show specific columns**
```sql
-- Show the population of Germany
SELECT population 
FROM world 
WHERE name = 'Germany'
```

**LIKE — pattern matching**
```sql
-- Show countries whose name starts with 'Al'
SELECT name 
FROM world 
WHERE name LIKE 'Al%'

-- % means 'anything can be here'
-- 'Al%' means starts with Al
-- '%land' means ends with land
-- '%ama%' means contains ama anywhere
```

**NOT LIKE — exclude a pattern**
```sql
-- Show countries whose name does NOT contain a space
SELECT name 
FROM world 
WHERE name NOT LIKE '% %'
```

**Tricky Question 13 — country with all vowels and no spaces**
```sql
-- Wrong approach: LIKE '%aeiou%' only finds vowels in exact order
-- Correct approach: separate LIKE for each vowel
SELECT name
FROM world
WHERE name LIKE '%a%'
  AND name LIKE '%e%'
  AND name LIKE '%i%'
  AND name LIKE '%o%'
  AND name LIKE '%u%'
  AND name NOT LIKE '% %'
```

### Day 2 — WHERE conditions, BETWEEN, IN, ORDER BY, LIMIT

**BETWEEN — filter a range**
```sql
-- Countries with area between 200,000 and 250,000 sq km
SELECT name, area
FROM world
WHERE area BETWEEN 200000 AND 250000
```

**IN — match a list of specific values**
```sql
-- Show Sweden, Norway, Denmark
SELECT name, population
FROM world
WHERE name IN ('Sweden', 'Norway', 'Denmark')
```

**AND — both conditions must be true**
```sql
-- Countries in Europe with population > 50 million
SELECT name
FROM world
WHERE continent = 'Europe' AND population > 50000000
```

**OR — either condition is true**
```sql
-- Countries in Europe OR Asia
SELECT name
FROM world
WHERE continent = 'Europe' OR continent = 'Asia'
```

**ORDER BY — sort results**
```sql
-- Show countries sorted by population (highest first)
SELECT name, population
FROM world
ORDER BY population DESC

-- ASC = ascending (lowest first), DESC = descending (highest first)
```

**LIMIT — restrict number of rows**
```sql
-- Show only the top 5 most populous countries
SELECT name, population
FROM world
ORDER BY population DESC
LIMIT 5
```

**ROUND — round numbers**
```sql
-- Show GDP per capita rounded to 2 decimal places
SELECT name, ROUND(gdp/population, 2) AS gdp_per_capita
FROM world
```

### Day 3 — GROUP BY, HAVING, Aggregate Functions

Aggregate functions summarize data — instead of showing every row, they calculate something across rows.

**The 5 aggregate functions**
```sql
SELECT COUNT(name)       -- count number of rows
SELECT SUM(population)   -- add up all values
SELECT AVG(gdp)          -- calculate average
SELECT MAX(area)         -- find highest value
SELECT MIN(population)   -- find lowest value
FROM world
```

**GROUP BY — split data into groups then aggregate**
```sql
-- How many countries are in each continent?
SELECT continent, COUNT(name)
FROM world
GROUP BY continent

-- Think of it like a Pivot Table in Excel
-- Split by continent, then count for each group
```

**HAVING — filter AFTER grouping (WHERE filters BEFORE)**
```sql
-- Show continents with total population > 500 million
SELECT continent, SUM(population)
FROM world
GROUP BY continent
HAVING SUM(population) > 500000000

-- Rule: WHERE comes before GROUP BY
--       HAVING comes after GROUP BY
```

**Combining WHERE and HAVING**
```sql
-- Total population per continent, excluding Antarctica,
-- only show continents with population > 100 million
SELECT continent, SUM(population)
FROM world
WHERE continent != 'Antarctica'
GROUP BY continent
HAVING SUM(population) > 100000000
```

**Finding the continent with highest total area**
```sql
-- Pattern: GROUP BY → ORDER BY → LIMIT 1
-- Use this whenever asked 'which category has the most/least'
SELECT continent, SUM(area)
FROM world
GROUP BY continent
ORDER BY SUM(area) DESC
LIMIT 1
```

### Day 5 — Pandas: Loading, Reading, Filtering

**SQL vs Pandas — same thing, different syntax**

| SQL | Pandas |
|-----|--------|
| `SELECT name FROM world` | `df['name']` |
| `WHERE population > 100` | `df[df['population'] > 100]` |
| `ORDER BY population DESC` | `df.sort_values('population', ascending=False)` |
| `LIMIT 5` | `df.head(5)` |
| `COUNT(*)` | `df.shape[0]` |
| `GROUP BY` | `df.groupby()` |

````python
# Creating a DataFrame manually
# A dictionary where keys = column names, values = list of data
data = {
    'Name': ['Rahul', 'Priya', 'Amit', 'Sara'],
    'Marks': [85, 90, 78, 92],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Chennai'],
    'Attendance': [90, 75, 85, 60]
}

df = pd.DataFrame(data)
df    # display the dataframe
````

````python
# Loading a CSV file
# df = pd.read_csv('matches.csv')   # uncomment and use your actual file path

# Basic exploration — always do these first when you load a new dataset
df.head()        # first 5 rows
# df.tail()      # last 5 rows
# df.shape       # (number of rows, number of columns)
# df.columns     # list of column names
# df.info()      # column names + data types + null count
# df.describe()  # statistical summary of numeric columns
````

````python
# Selecting columns
df['Name']                  # single column — returns a Series
# df[['Name', 'Marks']]     # multiple columns — returns a DataFrame
````

````python
# Filtering rows — most important Pandas skill

# Single condition (same as SQL WHERE)
df[df['City'] == 'Delhi']                    # rows where City is Delhi

# Note: use == for comparison, not = (single = is assignment in Python)
````

````python
# Multiple conditions
# Important: use & instead of AND, | instead of OR
# Always wrap each condition in () brackets

# AND condition — both must be true
df[(df['City'] == 'Delhi') & (df['Marks'] > 80)]

# OR condition — either must be true
# df[(df['City'] == 'Delhi') | (df['City'] == 'Mumbai')]
````

````python
# Sorting
df.sort_values('Marks', ascending=False)   # highest marks first
# df.sort_values('Marks', ascending=True)  # lowest marks first
````

````python
# Aggregate / summary functions
print(df['Marks'].mean())    # average marks
print(df['Marks'].sum())     # total marks
print(df['Marks'].max())     # highest marks
print(df['Marks'].min())     # lowest marks
print(df['Marks'].count())   # number of rows
````

````python
# groupby — Pandas version of SQL GROUP BY
# How many students from each city?
df.groupby('City')['Name'].count()
````

### loc vs iloc — how to select specific rows and columns

- `loc` — select by **label** (name)
- `iloc` — select by **position number** (0, 1, 2...)

````python
# loc — uses label names
df.loc[0]                        # row with label 0 (Rahul)
# df.loc[0, 'Name']              # row 0, column 'Name' → 'Rahul'
# df.loc[0:2, 'Name':'Marks']    # rows 0-2, columns Name to Marks
````

````python
# iloc — uses position numbers only, never names
df.iloc[0]           # first row (position 0)
# df.iloc[0, 0]      # row position 0, column position 0 → 'Rahul'
# df.iloc[-1]        # last row
# df.iloc[0:3]       # first 3 rows
````

````python
# When loc and iloc feel the same:
# That's because our index labels are 0,1,2,3 — same as positions
# They differ when index labels are not numbers

# Example — fruit_sales exercise from Kaggle
fruit_sales = pd.DataFrame({
    'Apples': [35, 41],
    'Bananas': [21, 34]
}, index=['2017 Sales', '2018 Sales'])   # note: capital S in Sales

fruit_sales.loc['2017 Sales']    # uses label name — works
# fruit_sales.iloc[0]            # uses position 0 — also works
# fruit_sales.iloc['2017 Sales'] # ERROR — iloc only accepts numbers
````

### map() and apply() — transforming column values

````python
# map() — transforms every value in a single column
# Works on ONE column at a time

# Way 1: using a dictionary (direct lookup)
city_map = {
    'Delhi': 'North',
    'Mumbai': 'West',
    'Chennai': 'South'
}
df['Region'] = df['City'].map(city_map)
df
````

````python
# Way 2: using a function
def get_grade(mark):
    if mark >= 90:
        return 'A'
    elif mark >= 80:
        return 'B'
    else:
        return 'C'

df['Grade'] = df['Marks'].map(get_grade)
df
````

````python
# Way 3: using a lambda (one-line shortcut function)
# lambda x: means 'for each value, call it x, then do this'
df['Marks_out_of_10'] = df['Marks'].map(lambda x: round(x / 10, 1))
df
````

````python
# apply() — more powerful than map
# Can use MULTIPLE columns at once — map cannot do this

# axis=1 means 'apply function to each ROW'
# axis=0 means 'apply function to each COLUMN' (default)

def get_result(row):
    # This function receives an entire row, not just one value
    if row['Marks'] > 80 and row['Attendance'] > 80:
        return 'Pass'
    else:
        return 'Fail'

df['Result'] = df.apply(get_result, axis=1)
df
````

````python
# apply() with lambda — combining two columns in one line
df['Total_score'] = df.apply(lambda row: row['Marks'] + row['Attendance'], axis=1)
df
````

````python
# map vs apply — when to use which
# Use map when:  transforming one column, or using a dictionary
# Use apply when: logic needs two or more columns
````

### idxmax() and idxmin()

````python
# idxmax() — returns the INDEX of the row with the HIGHEST value
# idx = index, max = maximum

df['Marks'].idxmax()    # returns the index label of the highest mark

# To get the full row of the topper:
topper_index = df['Marks'].idxmax()
df.loc[topper_index]
````

````python
# One-liner version
df.loc[df['Marks'].idxmax()]    # full row of topper

# idxmin() — same but for the LOWEST value
df.loc[df['Marks'].idxmin()]    # full row of lowest scorer
````
